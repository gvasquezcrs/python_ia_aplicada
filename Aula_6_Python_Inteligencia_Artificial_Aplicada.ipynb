```python
respuestas_cat = client_groq.chat.completions.create(
    model="llama-3.3-70b-versatile",
    messages=[
        {
                "role": "user",
                "content": f"""Eres un analista de datos, te pasaré muchas reseñas separadas por "####"
                Quiero que encuentres *exactamente* 5 categorias diferentes para cada tipo de reclamo.
                Las categorias deben ser solo una palabra.
                Despues quiero que me retornes la respuesta en formato JSON con tres claves:
                - "review_original": Aqui va la review original;
                - "review_es": Reseña traducida al español;
                - "categoria": Categoria definida correspondiente a esa reseña.

                Ejemplo:

                {{"review_original": "I didn´t like the color of the product"}}

                La respuesta debe ser SOLO y UNICAMENTE el texto en JSON

                          Reseñas: {reviews_negativas_unidas}"""
            }
       ],
    response_format={"type": "json_object"},
    max_tokens=4096
    )
json_de_reviews_clasificadas = respuestas_cat.choices[0].message.content
print(json_de_reviews_clasificadas)
```
