```python
import pandas as pd
import random

# Definir posibles valores para las columnas
nombres_productos_posibles = [
    'Laptop', 'Smartphone', 'Tablet', 'Smartwatch', 'Auriculares', 'Teclado', 'Mouse', 'Monitor', 'Impresora', 'Altavoz',
    'Cámara', 'Disco Duro', 'Router', 'Webcam', 'Micrófono', 'Gafas VR', 'Consola', 'Control Remoto', 'Proyector', 'Batería Externa'
]
categorias_posibles = ['Electrónica', 'Hogar', 'Oficina', 'Entretenimiento', 'Accesorios']

# Seleccionar 10 nombres de productos únicos aleatoriamente
productos_elegidos = random.sample(nombres_productos_posibles, 10)

# Generar datos para las 10 filas
data = {
    'Nombre de producto': productos_elegidos,
    'Categoria': [random.choice(categorias_posibles) for _ in range(10)],
    'Precio': [round(random.uniform(50, 1500), 2) for _ in range(10)],
    'Stock': [random.randint(10, 100) for _ in range(10)]
}

# Crear el DataFrame
df_productos = pd.DataFrame(data)

# Mostrar el DataFrame
display(df_productos)
```

```python
import os
from google.colab import userdata
from groq import Groq

os.environ["GROQ_API_KEY"] = userdata.get('GROQ_API_KEY')

client_groq = Groq()
```

```python
import pandas as pd
```

```python
#Etapa 1
df_reviews=pd.read_csv("/content/reviews.csv")
```

```python
df_reviews.head()
```

```python
columna_de_reviews = df_reviews["reviewText"]
```

```python
columna_de_reviews
```

```python
#Etapa 2

lista_de_analisis=[]
for numero, review in enumerate(columna_de_reviews):
    respuesta = client_groq.chat.completions.create(
        model="llama-3.3-70b-versatile",
        messages=[
                {
                    "role": "user",
                    "content": f""" Debes analizar la reseña que yo te envio y retornar un analisis de sentimiento. Debes responder
                    con: Positiva, Negativa o Neutra. Ejemplo:

                    Me encanto el producto -> Positiva
                    El producto no es nada especial -> Neutra
                    No me gusto el producto -> Negativa

                    Debes responder solamente una de esas tres opciones (Positiva, Neutra o Negativa) No agregues texto o explicaciones.

                    Reseña: {review}"""

                }

            ]
        )
    texto = respuesta.choices[0].message.content

    print(f"Review {numero}: {review} -> Sentimiento: {texto}")
    lista_de_analisis.append(texto)
```

```python
lista_de_analisis
```

```python
#Etapa 3

df_reviews["Sentimiento"] = lista_de_analisis

df_reviews.head()
```
