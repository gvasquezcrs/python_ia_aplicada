```python
salones = ["Salón 1", "Salón 2","Salón 3"]
```

```python
import random

def ubica_alumnos(nombre_alumno,lista_salones):
  salon_asignado=random.choice(lista_salones)

  dict_alumnos= {
      "nombre":nombre_alumno,
      "salon":salon_asignado
  }
  print(dict_alumnos)
```

```python
def corrije_texto(texto):
  texto_nuevo=texto.upper()
  lista_palabras=texto_nuevo.split()
  nombre_final= " ".join(lista_palabras)
  return nombre_final
```

```python
nombre_optimizado=corrije_texto(nombres_2)

ubica_alumnos(nombre_optimizado,salones)
```

```python
ubica_alumnos("Roberto Gonzalez",salones)
```
